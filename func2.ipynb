Write a function that takes a single movie and returns True if its IMDB score is above 5.5

```python
def score(movie): 
    return movie["imdb"]>5.5

print(score({
"name": "The Choice",
"imdb": 6.2,
"category": "Romance"
}))
print(score({
"name": "Bride Wars",
"imdb": 5.4,
"category": "Romance"
}))
```

Write a function that returns a sublist of movies with an IMDB score above 5.5.

```python
def highscore(movies):
    result = []
    for film in movies:
        if film["imdb"] > 5.5:
            result.append(film["name"])
    return result

movies=[{
"name": "Usual Suspects", 
"imdb": 7.0,
"category": "Thriller"
},
{
"name": "Hitman",
"imdb": 6.3,
"category": "Action"
},
{
"name": "Dark Knight",
"imdb": 9.0,
"category": "Adventure"
},
{
"name": "The Help",
"imdb": 8.0,
"category": "Drama"
}]
print(highscore(movies))
```
